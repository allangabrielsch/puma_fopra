```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

# Load data
data_3 = pd.read_csv('p5720_00008370.dat', delim_whitespace=True, comment='#')

h_3 = data_3['h']
det2_3 = data_3['det2']
error_3 = np.sqrt(det2_3)

# Define Gaussian function
def gaussian(x, amp, cen, wid, b):
    return amp * np.exp(-(x-cen)**2 / (2*wid**2)) + b

# Subset data for the first Gaussian (1.6 to 2.0)
mask1 = (h_3 >= 1.6) & (h_3 <= 2.0)
h_3_1 = h_3[mask1]
det2_3_1 = det2_3[mask1]

# Initial guess for the first Gaussian
initial_guess_1 = [max(det2_3_1), 1.92, np.std(h_3_1), min(det2_3_1)]

# Fit the first Gaussian with bounds
params_1, _ = curve_fit(gaussian, h_3_1, det2_3_1, p0=initial_guess_1, bounds=([0, 1.6, 0, 0], [np.inf, 2.0, np.inf, np.inf]), maxfev=20000)

# Subset data for the second Gaussian (2.0 to 2.4)
mask2 = (h_3 >= 2.0) & (h_3 <= 2.4)
h_3_2 = h_3[mask2]
det2_3_2 = det2_3[mask2]

# Initial guess for the second Gaussian
initial_guess_2 = [max(det2_3_2), 2.05, np.std(h_3_2), min(det2_3_2)]

# Fit the second Gaussian with bounds
params_2, _ = curve_fit(gaussian, h_3_2, det2_3_2, p0=initial_guess_2, bounds=([0, 2.0, 0, 0], [np.inf, 2.4, np.inf, np.inf]), maxfev=20000)

# Generate fitted data
gaussian1_data = gaussian(h_3, *params_1)
gaussian2_data = gaussian(h_3, *params_2)

# Create the plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot the original data with error bars
ax1.errorbar(h_3, det2_3, yerr=error_3, fmt='.', linestyle='-', color='b', ecolor='b', capsize=5, label='Data', alpha=0.5)

# Plot individual Gaussian fits
ax1.plot(h_3, gaussian1_data, 'g--', label='Gaussian 1 (1.6 to 2.0)')
ax1.plot(h_3, gaussian2_data, 'm--', label='Gaussian 2 (2.0 to 2.4)')

# Add labels and title
ax1.set_xlabel('H in (H,H,0) [r.l.u.]')
ax1.set_ylabel('Counts (a.u.)')
ax1.legend()
ax1.grid(True)


plt.show()

# Print the peaks and amplitudes of both Gaussians
peak_1, amplitude_1 = params_1[1], params_1[0]
peak_2, amplitude_2 = params_2[1], params_2[0]
print(f"Peak of Gaussian 1: {peak_1}")
print(f"Peak of Gaussian 2: {peak_2}")

peaks_x=[peak_1, peak_2]
peaks_y=[3,3]



```

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib.patches import Patch

h = [h_3, h_5, h_7, h_9]
det2 = [det2_3, det2_5, det2_7, det2_9]
x_values = [3, 5, 7, 9]

fig = plt.figure(figsize=(10, 12))
ax = fig.add_subplot(111, projection='3d')

# Plot data for each x value and connect the points
for i in range(len(x_values)):
    x = x_values[i]
    data_x = det2[i]
    h_x = h[i]
    errors = np.sqrt(data_x)  # Calculate error bars as sqrt(n)
    
    # Plot the main curve with error bars
    main_curve, = ax.plot([x] * len(h_x), h_x, data_x, marker='', linestyle=':', linewidth=3, label=f'x={x}')
    ax.add_collection3d(plt.fill_between(h_x, data_x - np.sqrt(data_x), alpha=0.25, color=main_curve.get_color()), zs=x, zdir='x')
    ax.add_collection3d(plt.fill_between(h_x, data_x - np.sqrt(data_x), data_x + np.sqrt(data_x), alpha=0.5, color=main_curve.get_color()), zs=x, zdir='x')

# Add titles and labels with reduced font size
ax.set_title('2D Spectra at different Q Values', fontsize=14)
ax.set_xlabel('Q value [meV]', fontsize=12)
ax.set_ylabel('H in (H,H,0) [r.l.u]', fontsize=12)
ax.set_zlabel('Counts [a.u]', fontsize=12)

# Create a proxy artist for the uncertainty shades
shade_patch = Patch(color='gray', alpha=0.5, label='Uncertainty')

# Add grid and set background color
ax.grid(True)

# Adjust view angle
ax.view_init(elev=20., azim=-35)

plt.tight_layout()

# Adjust the margins to prevent the z-axis label from getting cropped
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.2)

plt.show()

fig.savefig('3D_Spectra.pdf', format='pdf')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

h = [h_3, h_5, h_7, h_9]
det2 = [det2_3, det2_5, det2_7, det2_9]
x_values = [3, 5, 7, 9]

fig, ax = plt.subplots(figsize=(12, 8))

# Plot data for each x value and connect the points
for i in range(len(x_values)):
    x = x_values[i]
    data_x = det2[i]
    errors = np.sqrt(data_x)  # Calculate error bars as sqrt(n)
    h_x = h[i]
    main_curve, = ax.plot(h_x, data_x, marker='o', linestyle='-', label=f'x={x}')
    ax.errorbar(h_x, data_x, yerr=errors, fmt='.', color=main_curve.get_color(), capsize=5)
    
    # Shade under the curve
    ax.fill_between(h_x, data_x, alpha=0.3)

# Add titles and labels
ax.set_title('Spectrum of K$_{0.8}$Fe$_{1.7}$Se$_2$ at different $Q$ values', fontsize=16)
ax.set_xlabel('H in (H,H,0) [r.l.u]')
ax.set_ylabel('Counts [a.u]')

# Customize legend
ax.legend(title='Q Values [meV]', loc='best')

# Add grid and set background color
ax.grid(True)
ax.set_facecolor('whitesmoke')

plt.savefig('spectrum2D.pdf', format='pdf')
plt.show()
```
